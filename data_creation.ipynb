```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
num_samples = 12000
num_features = 55

financial_features = [
    "Revenue", "Cost of Goods Sold", "Gross Profit", "Operating Expenses", "EBITDA",
    "Net Income", "Total Assets", "Total Liabilities", "Equity", "Operating Cash Flow",
    "Investing Cash Flow", "Financing Cash Flow", "Current Assets", "Current Liabilities",
    "Working Capital", "Debt to Equity Ratio", "Return on Assets", "Return on Equity",
    "Revenue Growth", "Net Profit Margin", "EBITDA Margin", "Interest Coverage Ratio",
    "Earnings per Share", "Book Value per Share", "Dividend Yield", "Price to Earnings Ratio",
    "Price to Book Ratio", "Free Cash Flow", "Capital Expenditures", "Inventory Turnover",
    "Accounts Receivable Turnover", "Accounts Payable Turnover", "Days Sales Outstanding",
    "Days Payable Outstanding", "Days Inventory Outstanding", "Asset Turnover",
    "Current Ratio", "Quick Ratio", "Cash Ratio", "Debt Ratio", "Gross Margin",
    "Operating Margin", "Net Profit Growth", "ROIC", "EBIT", "NOPAT", "Total Debt",
    "Short-term Debt", "Long-term Debt", "Market Capitalization", "Enterprise Value",
    "Dividend Payout Ratio", "Interest Expense", "Tax Rate"
]
```

```python
financial_data = np.random.normal(loc=0, scale=1, size=(num_samples, len(financial_features)))
```

```python
df = pd.DataFrame(financial_data, columns=financial_features)
```
